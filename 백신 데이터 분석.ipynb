# 백신 접종 데이터 불러오기 및 전처리

```python
vaccine_path = '/content/drive/MyDrive/DATATHON01/covid_vaccine.csv'
vaccine_data = pd.read_csv(vaccine_path, encoding='CP949')
vaccine_data
```

```python
vaccine_data = vaccine_data.sort_values(['접종일'])
vaccine_data
```

```python
# 중복데이터 제거
vaccine_data.drop_duplicates(inplace=True)
```

```python
# 1차 접종자 데이터
vaccine_1th = vaccine_data[['접종일', '당일 1차접종자 수', '1차접종 누계', '1차접종률(%)']]
vaccine_1th
```

```python
# 2차 접종자 데이터
vaccine_2th = vaccine_data[['접종일', '당일 2차접종자 수', '2차접종 누계', '2차접종률(%)']]
vaccine_2th
```

```python
# 1차 접종 데이터 컬럼명 변경
vaccine_1th.rename(columns = {'접종일' : 'date'}, inplace=True)
vaccine_1th
```

```python
# 2차 접종 데이터 컬럼명 변경
vaccine_2th.rename(columns = {'접종일' : 'date'}, inplace=True)
vaccine_2th
```

```python
# 1,2차 접종자 데이터
vaccine_both = pd.merge(vaccine_1th, vaccine_2th, on='date')
vaccine_both
```

```python
# 결측치 확인하기
vaccine_1th.isnull().sum()
```

```python
vaccine_2th.isnull().sum()
```

# 전국 코로나 확진자 데이터 불러오기

```python
covid_path = '/content/drive/MyDrive/DATATHON01/confirmed_case.csv'
covid_data = pd.read_csv(covid_path, encoding='CP949')
covid_data
```

```python
# 전국 데이터만 추출
covid_data = covid_data[['전국 기준일', '전국 확진', '전국 추가 확진', '전국 치료중', '전국 퇴원', '전국 사망']]
covid_data
```

```python
# 컬럼명 변경
covid_data.rename(columns = {'전국 기준일' : 'date'}, inplace=True)
covid_data
```

```python
# 중복 데이터 제거
covid_data.drop_duplicates(inplace=True)
```

```python
# 결측치 확인하기
covid_data.isnull().sum()
```

```python
# 전국 코로나 확진자 데이터와 백신 데이터 기간 맞추기
# 21.04.21 ~ 22.03.06

covid_data = covid_data[covid_data['date']>='2021-04-21']
covid_data = covid_data[covid_data['date']<='2022-03-06']
covid_data
```
